```python
import pandas as pd
import sqlite3

def carregar_dados():
    # Conectar ao banco de dados
    conn = sqlite3.connect('/home/selaranjeira/Desktop/loc_proj/desemp_esportivo/data/runningDB.sqlite3')
    
    # Verificar a conexão e a existência da tabela
    query_tables = "SELECT name FROM sqlite_master WHERE type='table';"
    tables = pd.read_sql_query(query_tables, conn)
    print("Tabelas no banco de dados:", tables)

    # Carregar os dados da tabela running_data
    try:
        df = pd.read_sql_query("SELECT * FROM running_data", conn)
        print("Dados carregados do banco de dados:")
        print(df.head())  # Mostra as primeiras linhas do DataFrame

        # Fechar a conexão
        conn.close()

        # Processar os dados
        df['data'] = pd.to_datetime(df['data'])
        df['quilometragem'] = df['quilometragem'].astype(float)
        return df
    except Exception as e:
        print("Erro ao carregar os dados:", e)
        conn.close()
        return pd.DataFrame()  # Retornar um DataFrame vazio em caso de erro

# Testar a função
df = carregar_dados()
print(df.head())
```

```python
# Função para criar gráfico de meia rosca
def criar_grafico_meia_rosca(calorias_corrente, calorias_anterior, titulo):
    if calorias_corrente > calorias_anterior:
        # Calorias correntes ultrapassam a meta
        fig = go.Figure()

        fig.add_trace(go.Pie(
            values=[calorias_anterior, calorias_corrente - calorias_anterior],
            labels=['Meta Anterior', 'Excesso de Calorias'],
            hole=.4,
            direction='clockwise',
            sort=False,
            marker=dict(colors=['#dddddd', '#ff6361']),
            showlegend=False
        ))

        fig.update_layout(
            title=titulo,
            annotations=[
                dict(text=f'{calorias_corrente:.0f}', x=0.5, y=0.5, font_size=20, showarrow=False),
                dict(text='Meta Superada!', x=1.1, y=1.1, font_size=15, showarrow=False, font_color="green")
            ],
            width=450,
            height=450
        )
    else:
        # Calorias correntes dentro da meta
        fig = go.Figure()

        fig.add_trace(go.Pie(
            values=[calorias_corrente, calorias_anterior - calorias_corrente],
            labels=['Calorias Correntes', 'Meta Restante'],
            hole=.4,
            direction='clockwise',
            sort=False,
            marker=dict(colors=['#ff6361', '#dddddd']),
            showlegend=False
        ))

        fig.update_layout(
            title=titulo,
            annotations=[
                dict(text=f'{calorias_corrente:.0f}', x=0.5, y=0.5, font_size=20, showarrow=False),
                dict(text='Meta Aquém', x=1.1, y=1.1, font_size=15, showarrow=False, font_color="red")
            ],
            width=450,
            height=450
        )

    return fig

# Converter coluna de data para datetime
df['data'] = pd.to_datetime(df['data'], format='%d/%m/%y')

# Calcular calorias mensais
calorias_mensais = df.groupby(df['data'].dt.to_period('M'))['calorias'].sum()

# Criar gráficos para três meses
figures = []
for i in range(len(calorias_mensais)):
    mes_corrente = calorias_mensais.index[i]
    mes_anterior = calorias_mensais.index[i-1] if i > 0 else mes_corrente
    
    calorias_corrente = calorias_mensais[mes_corrente]
    calorias_anterior = calorias_mensais[mes_anterior]
    
    titulo = f'Meta Calorias Mensal ({mes_corrente})'
    fig = criar_grafico_meia_rosca(calorias_corrente, calorias_anterior, titulo)
    figures.append(fig)

# Exibir os gráficos no Jupyter Notebook
for fig in figures:
    fig.show()

```
